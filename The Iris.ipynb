# Modeling - Simple manual model
after manual analysis through graphs and EDA don before we get a knowledge that any flower with petal length less than 2~2.5 cm is 100% a setosa flower

```python
#creating a single feature function
def single_feature_prediction(petal_length):
    if(petal_length <2.5):
        return 0 #meaning setosa's target number
    elif(petal_length<4.75):
        return 1
    else:
        return 2
```

Above we just added some if else statements roughly by seeing the graphs

```python
df_train.columns
```

```python
#this one will only give the rows of the second column as x_train[--this means rows--,--this means column]
x_train[:,2]
```

```python
predictions=[]
for val in x_train[:,2]:
    predictions.append(single_feature_prediction(val))
```

```python
len(predictions)
```

```python
hit=0
for i in range(0,112):
    if(predictions[i]==y_train[i]):
        hit=hit+1
print(hit/112)
```

**HERE WE CAN CLEARLY SEE** we got a solid _95.5%_ of accuracy with the oreiginal data just by seeing the graphs
